```python
import numpy as np
import matplotlib.pyplot as plt
import os
import sys

module_path = os.path.abspath(os.path.join(r'C:\Users\christopher.battle\Documents\ml_projects\grampyx_project'))
if module_path not in sys.path:
    sys.path.append(module_path)
    
import grampyx.grampyx as gpx
```

```python
s = "grampyx"

plt.figure(figsize=(10,7))
for i, mapping in enumerate(["ordered", "frequency", "aesthetic"]):
    plt.subplot(2,3,i+1)
    ax = plt.gca()
    plt.imshow(gpx._word2pic(s, mapping=mapping, pictype="punchcard"), cmap="gray", origin="lower")
    plt.title(mapping + " punchcard")
    ax.axes.get_xaxis().set_visible(False)
    ax.axes.get_yaxis().set_visible(False)
    plt.subplot(2,3,i+4)
    ax = plt.gca()
    plt.imshow(gpx._word2pic(s, mapping=mapping), cmap="gray", origin="lower")
    plt.title(mapping + " gradient")
    ax.axes.get_xaxis().set_visible(False)
    ax.axes.get_yaxis().set_visible(False)
```
